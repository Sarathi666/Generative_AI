```python
!pip install transformers torch
from transformers import pipeline

# Load a pre-trained GPT-2 model (small version for local CPU use)
chatbot = pipeline("text-generation", model="gpt2", device=-1)  # device=-1 means CPU

def chatbot_response(user_input):
    # Generate a response from the model
    response = chatbot(user_input, max_length=100, num_return_sequences=1)

    # Extract the text of the response
    generated_text = response[0]['generated_text']

    # Return the response without the input prompt (trim the input prompt from the output)
    return generated_text[len(user_input):].strip()

# Example usage
user_input = "Tell me a joke!"
response = chatbot_response(user_input)
print(response)
# Load a pre-trained GPT-2 model (small version for local CPU use)
chatbot = pipeline("text-generation", model="gpt2", device=-1)  # device=-1 means CPU

def chatbot_response(user_input):
    # Generate a response from the model
    response = chatbot(user_input, max_length=100, num_return_sequences=1)

    # Extract the text of the response
    generated_text = response[0]['generated_text']

    # Return the response without the input prompt (trim the input prompt from the output)
    return generated_text[len(user_input):].strip()

# Example usage
user_input = "Tell me a joke!"
response = chatbot_response(user_input)
print(response)
```
