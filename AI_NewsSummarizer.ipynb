```python
# Install necessary libraries
!pip install gradio transformers

# Import libraries
import gradio as gr
from transformers import pipeline

# Initialize summarization pipeline
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Function to summarize news articles
def summarize_article(article_text):
    # Generate the summary
    summary = summarizer(article_text, max_length=200, min_length=50, do_sample=False)
    return summary[0]['summary_text']

# Create a Gradio interface
interface = gr.Interface(
    fn=summarize_article,                  # Function to call
    inputs=gr.Textbox(lines=10, label="Paste News Article Here"),  # Input type (textbox for article)
    outputs=gr.Textbox(label="Summary"),    # Output type (textbox for summary)
    title="News Summarization Tool",        # Title of the interface
    description="This tool takes a news article as input and provides a concise summary."
)

# Launch the interface
interface.launch()
```
