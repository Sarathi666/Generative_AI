```python
# ✅ Step 1: Install all dependencies
!pip install -q spacy pdfplumber
!python -m spacy download en_core_web_sm

# ✅ Step 2: Upload your resume (PDF or TXT)
from google.colab import files
import pdfplumber
import spacy

uploaded = files.upload()
file_name = list(uploaded.keys())[0]

# ✅ Step 3: Extract text from uploaded file
def extract_text(file_path):
    if file_path.endswith('.pdf'):
        with pdfplumber.open(file_path) as pdf:
            return "\n".join([page.extract_text() or "" for page in pdf.pages])
    else:
        with open(file_path, 'r', encoding='utf-8') as f:
            return f.read()

resume_text = extract_text(file_name)

# ✅ Step 4: Run NLP analysis
nlp = spacy.load("en_core_web_sm")
doc = nlp(resume_text)

# ✅ Step 5: Extract info
skills_keywords = [
    "python", "java", "sql", "javascript", "html", "css", "react", "node",
    "aws", "docker", "git", "c++", "c#", "machine learning", "nlp", "pandas"
]

entities = {
    "NAME": None,
    "EMAIL": [],
    "PHONE": [],
    "EDUCATION": [],
    "SKILLS": set(),
    "EXPERIENCE": []
}

import re

# Extract name (first PERSON entity)
for ent in doc.ents:
    if ent.label_ == "PERSON":
        entities["NAME"] = ent.text
        break

# Extract email
entities["EMAIL"] = re.findall(r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}", resume_text)

# Extract phone numbers
entities["PHONE"] = re.findall(r"\+?\d[\d\s\-]{8,}\d", resume_text)

# Extract education-related sentences
edu_keywords = ["bachelor", "master", "b.sc", "m.sc", "ph.d", "degree", "university", "college"]
for sent in doc.sents:
    if any(keyword in sent.text.lower() for keyword in edu_keywords):
        entities["EDUCATION"].append(sent.text.strip())

# Extract skills
for token in doc:
    if token.text.lower() in skills_keywords:
        entities["SKILLS"].add(token.text.lower())

# Extract experience-related lines
exp_keywords = ["experience", "worked", "project", "intern", "role", "responsible", "developed", "managed"]
for sent in doc.sents:
    if any(keyword in sent.text.lower() for keyword in exp_keywords):
        entities["EXPERIENCE"].append(sent.text.strip())

# ✅ Step 6: Display extracted info
from pprint import pprint
print("\n🔍 Resume Analysis:\n")
pprint(entities)
```
